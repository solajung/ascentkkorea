```python
import pandas as pd
from openpyxl import load_workbook
import os

directory = 'C:/Users/어센트코리아/Desktop/현황판/prime-kw-set'
all_data = pd.DataFrame()

for filename in os.listdir(directory):
    excel_file_path = os.path.join(directory, filename)
    wb = load_workbook(excel_file_path, read_only=True)  # read_only=True 추가
    ws = wb['Prime_kws_set']
    data = []
    for row in ws.iter_rows(min_row=9, min_col=2, values_only=True):
        data.append(row)
    df = pd.DataFrame(data[1:], columns=data[0])
    # 파일 이름에서 국가 코드 추출
    country_code = filename.split('_')[-1].split('.')[0]
    # 컬럼명에서 국가 코드 추출하여 괄호와 그 안의 문자열 제거
    df.columns = [col.split('(')[0].strip() if f" ({country_code})" in col else col for col in df.columns]
    # 국가 코드를 컬럼명에 추가
    df.columns = [col.replace(f" ({country_code})", "") if f" ({country_code})" in col else col for col in df.columns]
    # 'sitecode' 컬럼 추가
    df['sitecode'] = country_code
    all_data = pd.concat([all_data, df], ignore_index=True)

all_data.head()
```
