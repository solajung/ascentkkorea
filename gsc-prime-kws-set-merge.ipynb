# GSC + Prime KWs Set with Monthly Search Volume

```python
import pandas as pd
import glob

gsc_input_path = "C:/Users/어센트코리아/gsc-by-month/gsc-by-month-raw/*.csv"
gsc_file_list = [f for f in glob.glob(gsc_input_path)]

gsc_total = pd.DataFrame()

for file in gsc_file_list:
    temp = pd.read_csv(file, parse_dates = ['month'])
    gsc_total = pd.concat([gsc_total, temp])

gsc_total['sitecode'] = gsc_total['website'].str.split("/", expand=True)[3]

gsc_total = gsc_total.rename(columns = {"month": "Month", "sitecode": "Sitecode"})
gsc_total.to_csv('month-page-query-merge.csv', index=False)
gsc_total.info()
```

```python
df_gsc = pd.read_csv('month-page-query-merge.csv', parse_dates = ['Month'])

# prime kws set 불러와 병합하기
primekw_file_path = "C:/Users/어센트코리아/prime-kw-set/prime_kw_monthly_search_volume_excluded.csv"

prime_kws_set = pd.read_csv(primekw_file_path, parse_dates = ['Month'])
prime_kws_set['Sitecode'] = prime_kws_set['Sitecode'].str.lower()

gsc_primekw_merge = df_gsc.merge(prime_kws_set, how = 'outer', left_on = ['Sitecode', 'query', 'Month'], right_on = ['Sitecode', 'KEYWORDS', 'Month'])

gsc_primekw_merge["Global Search Volume"] = gsc_primekw_merge["Global Search Volume"].fillna(0)
gsc_primekw_merge["Global Search Volume"] = gsc_primekw_merge["Global Search Volume"].astype(int)
gsc_primekw_merge["Country Search Volume"] = gsc_primekw_merge["Country Search Volume"].fillna(0)
gsc_primekw_merge["Country Search Volume"] = gsc_primekw_merge["Country Search Volume"].astype(int)
#gsc_primekw_merge = gsc_primekw_merge.drop(columns=['Unnamed: 0'])

gsc_primekw_merge.to_csv('gsc_primekw_with_search_volume_merge.csv', index=False)
gsc_primekw_merge.info()
```
