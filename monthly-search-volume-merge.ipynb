```python
import pandas as pd
```

```python
global_data = pd.read_csv('global.csv')
global_data.head()
```

```python
country_data = pd.read_csv('country.csv')
country_data.head()
```

```python
global_melt = pd.melt(global_data, id_vars = ['sitecode', 'Keyword Set', 'Models', 'Strategic Topics', 'Rising Topics', 'CEJ', 'Channel', 'Page Type', 'Topic Depth1', 'Topic Depth2', 'Topic Depth3', 'Topic Depth4', 'Generic/Branded', 'KEYWORDS', 'Global Annual', 'Global Average Monthly', 'UK Annual', 'UK Average Monthly'], value_vars = ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01', '2023-09-01', '2023-10-01', '2023-11-01', '2023-12-01', '2024-01-01', '2024-02-01'], var_name='month', value_name='global search volume')
global_melt.head()
```

```python
country_melt = pd.melt(country_data, id_vars = ['sitecode', 'Keyword Set', 'Models', 'Strategic Topics', 'Rising Topics', 'CEJ', 'Channel', 'Page Type', 'Topic Depth1', 'Topic Depth2', 'Topic Depth3', 'Topic Depth4', 'Generic/Branded', 'KEYWORDS', 'Global Annual', 'Global Average Monthly', 'UK Annual', 'UK Average Monthly'], value_vars = ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01', '2023-09-01', '2023-10-01', '2023-11-01', '2023-12-01', '2024-01-01', '2024-02-01'], var_name='month', value_name='country search volume')
country_melt.head()
```

```python
global_country_merge = global_melt.merge(country_melt, how='left')
global_country_merge.head()
```

```python
global_country_merge.to_csv('prime_kw_monthly_search_volume.csv')
```
