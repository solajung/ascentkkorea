```python
import pandas as pd
import glob
```

```python
input_path = "C:/Users/USER/google-search-console/date-page-query/*.csv"
file_list = [f for f in glob.glob(input_path)]
file_list
```

```python
total = pd.DataFrame()

for file in file_list:
    temp = pd.read_csv(file, parse_dates=['date'])
    total = pd.concat([total, temp])

total['sitecode'] = total['website'].str.split("/", expand=True)[3]

kw_set = pd.read_csv('prime-keyword-set-raw-v1.0-20240402.csv')

total_merge_outer = total.merge(kw_set, how = 'outer', left_on = ['sitecode', 'query'], right_on = ['sitecode', 'KEYWORDS'])

total_merge_outer["Global Annual"] = total_merge_outer["Global Annual"].fillna(0)
total_merge_outer["Global Average Monthly"] = total_merge_outer["Global Average Monthly"].fillna(0)
total_merge_outer["Country Annual"] = total_merge_outer["Country Annual"].fillna(0)
total_merge_outer["Country Average Monthly"] = total_merge_outer["Country Average Monthly"].fillna(0)

total_merge_outer["Global Annual"] = total_merge_outer["Global Annual"].astype(int)
total_merge_outer["Global Average Monthly"] = total_merge_outer["Global Average Monthly"].astype(int)
total_merge_outer["Country Annual"] = total_merge_outer["Country Annual"].astype(int)
total_merge_outer["Country Average Monthly"] = total_merge_outer["Country Average Monthly"].astype(int)

total_merge_outer.to_csv('total_merge_outer.csv')

total_merge_outer.info()
```

```python
total_merge_left = total.merge(kw_set, how = 'left', left_on = ['sitecode', 'query'], right_on = ['sitecode', 'KEYWORDS'])
total_merge_left.to_csv('total_merge_left.csv')

total_merge_left.info()
```
